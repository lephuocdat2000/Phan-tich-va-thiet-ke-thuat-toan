```
import random
import numpy as np
import pandas as pd
import cProfile
#while True:
 # try:
  #  i = input()
   # t.append(i)
  #except:
   # break
def tree(key):
  r = len(key)
  m = r//2
  if r<1:
    return
  tree(key[:m])
  tree(key[m+1:])
  print(key[m])
  return
def Create_arr(n,arr):
  arr=np.zeros(n)
  for i in range(0,n):
    arr[i]=random.randint(0,100000)
  np.sort(arr)
  return arr
arr=[]
twodimen_arr=[]
for n in range(0,10000):
   arr=Create_arr(n,arr)
   runtime=cProfile.run('tree(arr)')
   twodimen_arr.append([n,runtime])
df=pd.DataFrame(data=twodimen_arr,columns=['n','runtime'])
```
